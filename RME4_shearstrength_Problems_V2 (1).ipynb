# **Barton-Bandis Shear Failure Criterion for Rock**






```python
# @title
import numpy as np
import matplotlib.pyplot as plt


def phi_residual(phiB,r,R):
  return (phiB-20) + 20*(r/R)
def BartonBandisShear(sigN,JRC,JCS,phiR):
  return sigN*np.tan((JRC*np.log10(JCS/sigN)+phiR)*np.pi/180)

JRC = 11
JCS = 100
phiR = 46

sigN0=10
sigN1=100000
sigN_inc=10
sigN_list = np.linspace(sigN0,sigN1,int((sigN1-sigN0)/(sigN_inc)))
shearBB_list = list([BartonBandisShear(sigN,JRC,JCS,phiR) for sigN in sigN_list])

#Plot
plt.figure(figsize=(10,6))
#plt.plot(sigN_list, shearBB_list, 'g-', linewidth = 1, label = "__")
plt.plot(sigN_list, list([BartonBandisShear(sigN,10,100,phiR) for sigN in sigN_list]), 'g-', linewidth = 1, label = "JRC=10 JCS=100 phiR=46")
plt.plot(sigN_list, list([BartonBandisShear(sigN,12,100,phiR) for sigN in sigN_list]), 'r-', linewidth = 1, label = "JRC=12 JCS=100 phiR=46")
plt.plot(sigN_list, list([BartonBandisShear(sigN,14,100,phiR) for sigN in sigN_list]), 'b-', linewidth = 1, label = "JRC=14 JCS=100 phiR=46")
plt.plot(sigN_list, list([BartonBandisShear(sigN,16,100,phiR) for sigN in sigN_list]), 'k-', linewidth = 1, label = "JRC=16 JCS=100 phiR=46")
plt.plot(sigN_list, list([BartonBandisShear(sigN,18,100,phiR) for sigN in sigN_list]), 'c-', linewidth = 1, label = "JRC=18 JCS=100 phiR=46")

plt.ylabel("Shear resistance [N]")
plt.xlabel("Normal load [N]")
plt.legend()
plt.grid(True)


```
