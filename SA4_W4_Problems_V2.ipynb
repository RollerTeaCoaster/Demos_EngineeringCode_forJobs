Looking at Fourier Expansions


```python
# @title
import numpy as np
import matplotlib.pyplot as plt



def pulseodd(A,n,T,t,xoffset,yoffset):
  return yoffset + (4*A)/((2*n-1)*np.pi)*np.sin((2*n-1)*(2*np.pi)/T*t - xoffset)

def pulseodd2(A,n,T,t,xoffset,yoffset):
  n_list = np.arange(1,n+1)
  f_a = 0
  for n in n_list:
    f_b = (4*A)/((2*n-1)*np.pi)*np.sin((2*n-1)*(2*np.pi)/T*t - xoffset)
    f_a+=f_b
  return yoffset + f_a

def pulseodd3(A,n,T,t,xoffset,yoffset):
  n_list = np.arange(1,n+1)
  f_a = 0
  for n in n_list:
    f_b = (2*A*((-1)**(n-1) + 1))/((2)*np.pi)*np.sin((n)*(2*np.pi)/T*t - xoffset)
    f_a+=f_b
  return yoffset + f_a

 #  yoffset + (4*A)/((2*n-1)*np.pi)*np.sin((2*n-1)*(2*np.pi)/T*t - xoffset)


def problem3baseexcitationsteadystatemotion(X_hat,forcingfreq,t,forcingphasediff):
  return X_hat * np.sin(forcingfreq*t - forcingphasediff)
def problem3baseexcitationsteadystateamplitude(Y,freqratio,dampingratio):
  return Y * ((1 + (2*freqratio*dampingratio)**2)/((1-freqratio**2)**2 + (2*freqratio*dampingratio)**2))**0.5

forcingfreq = 5*(2 * np.pi)
forcingamplitude = 0.01

m_eff = 10
k_eff = 1000
c_eff = 15

naturalfreq = (k_eff/m_eff)**0.5

freqratio = forcingfreq/naturalfreq
dampingratio = c_eff / (2 * (k_eff*m_eff)**0.5)

X_hat = problem3baseexcitationsteadystateamplitude(forcingamplitude,freqratio,dampingratio)
forcingphasediff = np.arctan((2*freqratio*dampingratio)/(1-freqratio**2)) + np.pi

t0=0
t1=2*np.pi
t_inc=np.pi/64
t_list = np.linspace(t0,t1,int((t1-t0)/(t_inc)))
displacement_baseexcitationsteadystate_list = list([problem3baseexcitationsteadystatemotion(X_hat,forcingfreq,t,forcingphasediff) for t in t_list])


n=5
n_list = np.arange(1,n+1)
print(n_list)
A=1
T=2

force_pulsefourierseries_list = []
force_pulsefourierseries_list = list([pulseodd2(A,n,T,t,0,0) for t in t_list])
force_pulsefourierseries_list_TRYOUT = list([pulseodd3(A,n,T,t,0,0) for t in t_list])


#Plot
plt.figure()

plt.plot(t_list, force_pulsefourierseries_list, 'g-', linewidth = 1, label = "__")

plt.plot(t_list, force_pulsefourierseries_list, 'r-', linewidth = 1, label = "__")


plt.legend()
plt.grid(True)



```
